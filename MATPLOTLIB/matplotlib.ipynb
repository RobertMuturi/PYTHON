It is an excellent 2D and 3D graphics library for generating scientific figures.

* Generally easy to get started for simple plots

Major set back

### **Modify Line Style**

There are two ways of modifying the linestyle. 

The first uses a format string called **fmt** that has a colour, a line style and a marker as the values. So to change the marker, type and colour of a line, you would use, as an example, k--x. This makes the colour black, the line type a dashed line, and have x as a marker on the various data points. To get what the different possibilities are for **fmt**, check out https://bit.ly/2UXEr6T

```python
plt.figure(figsize = (10,15))

plt.title('Loan Amount by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.plot(sector, loan, 'c--^')#modify line in the plot function

plt.show()
```

### Modifying Line Style 2
The second is a less confusing way of dealing with line types because it involves calling the characteristic directly by use of keyword arguements.

```python
plt.figure(figsize = (15, 5))#Default size is 6 by 4 inches

plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
plt.yticks(rotation = 60)

plt.ylabel('Loan Amount', fontsize = 15)  
plt.xlabel('Sector', fontsize = 15)

plt.plot(sector, loan, color = 'red', linestyle = ':', marker = 'o')
plt.show()
```

### **Add a legend**

The legend tells what each line represents. Just like a key on a map, the legend helps the reader understand what they are looking at.

We can use the legend() method where we specify the labels for each line by passing them as a list of arguements.
For this, the order in which the labels are passed has to match the order in which the graphs have been plotted. Changing the order will result in wrongly labelled variables.

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Lender Count by Sector', fontsize = 15)

plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.plot(sector, loan, c = '#EE66CC', ls = ':', marker = 'x')#loan amount
plt.plot(sector, lender, c = '#22ff77')#add another plt.plot function/ lender account

plt.legend(['Loan Amount', 'Lender Count'])

plt.show()
```

A better way is to include the `label` parameter in the `plt.plot()` method for each plot, then just call the `plt.legend()` method without any arguements. This way, it doesn't matter what order the graphs have been plotted, the legend will always point to the correct variable.

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Lender Count by Sector', fontsize = 15)

plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.plot(sector, loan, c = '#EE66CC', ls = ':', marker = 'x', label = 'Loan Amount')
plt.plot(sector, lender, c = '#22ff77', label = 'Lender Count')#add another plt.plot function

plt.legend()

plt.show()
```

### Fix: Dodge the bars, meaning, show them side by side Using Numpy

One thing to note about multiple barcharts is that they are generated in a overlapping format.

The last plot that is generated, is the one that appears above.

Funded amount is less than loan amount, so we are unable to see what is behind...

To do this, we have to play with the width and the positioning of the bars as shown in the next cell. 

1. We want to reduce the width of the bars so that we can be able to fit them side by side on the x axis.
2. Then shift the position of one bar by the width to be able to display the two bars side by side.

We generate an array of x labels using the np.arange(len()) method. This turns the categorical values on the x varaible(sector column) into an array of numbers which we can use to position the minibars on the x axis.

By default the width of bars is 0.8. We set the width  of the 2 Y variables to a smaller size (0.3), then again shift one of the Y variables by that width, leaving the other in place. This gives us the result we want.

```python
x_indices = np.arange(len(sector))
x_indices
```

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Funded Amount by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

# converting sector variable to get the position of ticks to insert on the x axis, 
#This array is going to be used in place of the categorical variable sector to set the position.
x_indices = np.arange(len(sector))

width = 0.3# specifying the width of the 2 bars

plt.bar(x_indices, loan, width, label = 'Loan Amount')
plt.bar((x_indices + width), fund, width, label = 'Funded Amount')

#Reset the xtick labels to the sector categories
plt.xticks(ticks = x_indices, labels = sector, rotation = 75)

plt.legend()

plt.show()
```

## Histogram

Appropriate to show the distribution of a variable.

By default it returns a plot with 10 bins. Bins are consecutive, non-overlapping intervals of a variable(represented by the bars in a histogram). The number of variables is divided into a series of intervals. A count is then made of how many values fall into each interval.

if you have a list of numbers 23, 25, 28, 9, 3, 7, 33, 1, you can have 4 bins from 1 to 10, 11 to 20, 21 to 30, 31 to 40. From the 0 to 10 bin there are 4 values, 11 to 20 0 values, 21 to 30 3 values, and 31 to 40 1 value.

```python
plt.hist(df['count'])
plt.plot()
```

```python
plt.figure(figsize = (10,5))

plt.title('Distribution of count', fontsize = 15)

plt.hist(df['count'], edgecolor = 'k', bins = 10)#bins are the towers

plt.show()
```

## Multiple Plots

Instead of having lots of information on one plot, we can choose to have multiple plots on one canvas, each showing one bit of information. Whether to use the multiple bits of information in one plot, or one bit of information per plot is a decision to be made based on the output/audience/data.

```python
plt.figure(figsize = (15,5))# create a canvas and specify its size.

plt.subplot(1,2,1)#1 row 2 columns and postion of the subplot.
#first we subdivide the canvas and specify the postion of the subplot.
plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
plt.xlabel('Sector')
plt.ylabel('Loan Amount')
plt.plot(sector, loan)

plt.subplot(1,2,2) 
plt.title('Lender Count by Sector')
plt.xticks(rotation = 75)
plt.xlabel('Sector')
plt.ylabel('Lender Count')
plt.plot(sector, lender)

plt.show()
```

```python
plt.figure(figsize = (20,20))
plt.subplot(2,2,1)# 2rows and 2 columns: first we subdivide the canvas and specify the postion of the subplot.
plt.title('Loan Amount by Sector')

plt.xticks(rotation = 45)

plt.plot(sector, loan)


plt.subplot(2,2,2)
plt.title('Distribution of Term in Months')

plt.xticks(rotation = 45)

plt.hist(df['count'], edgecolor = 'k', bins = 5)


plt.subplot(2,2,3)
plt.title('Lender Count by Sector')

plt.bar(sector, lender)


plt.subplot(2,2,4)
colours = np.arange(len(sector))

plt.title('Loan Amount vs Lender Count')

plt.scatter(loan, lender, c = colours, cmap = 'Blues',
            marker = 'o', edgecolor = 'k', alpha = 0.75, s = loan/1000)


plt.savefig('plot.png')#has to be the second last command before you call the plot.
plt.show()
```
