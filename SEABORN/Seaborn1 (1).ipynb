## Categorical Plots


There are several types. For this class we shall only look at the `barplot()`. For more, check out the documentation at https://seaborn.pydata.org/api.html (for the APIs) and https://seaborn.pydata.org/tutorial/categorical.html (for tutorials)

Bar plots help you visualize the distributions of categorical variables.

In seaborn there are two ways you can pass plotting variables.

* The first way (recommended) is to pass your DataFrame to `data=` the arguments,`x=` and `y=` while passing column names to the axes arguments, and .

```python

sns.barplot(x = 'sector', y = 'loan_amount', data = df)

plt.show()
```

* The second way is to directly pass in Series of data to the axes arguments.

```python

sns.barplot(x = df['sector'], y = df['loan_amount'])

plt.show()
```

Since we are not working with sample data we can remove the error bars.

Use the `ci` parameter set to `None` to remove the error bars. 

Error bars are graphical representations of the variability of data and used on graphs to indicate the error or uncertainty in a reported measurement(Confidence interval)

```python

sns.barplot(x = 'sector', y = 'loan_amount', data = df, ci = None)

plt.show()
```

We are going to use Matplotlib's functions to customize our plot.

`plt.figure(figsize = )`  to resize the plot

`plt.title()` to add a title

`plt.xticks()` method to rotate the labels on the x axis

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount by Sector')

plt.xticks(rotation = 75)

sns.barplot(x = 'sector', y = 'loan_amount', data = df, ci = None)

plt.show()
```

Add labels to the bars

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount by Sector')

plt.xticks(rotation = 75)

ax = sns.barplot(x = 'sector', y = 'loan_amount', data = df, ci = None)

#put values at the top of the bars
for p in ax.patches:
    ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
             ha = 'center', va = 'center', rotation = 60, xytext = (0, 20), textcoords = 'offset points')
    
plt.show()
```

### Box Plot

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

A box plot is composed of a summary of 5 different data points: the minimum, first quartile, median, third quartile, and maximum.

    Minimum
    First Quartile or 25%
    Median (Second Quartile) or 50%
    Third Quartile or 75%
    Maximum

```python
sns.boxplot(x="sector", y="average_count", data=df)
```

* The bottom black horizontal line of the box plot is the minimum value
* The first black horizontal line of the rectangle shape of the box plot is the first quartile
     or 25%
* The second black horizontal line of the rectangle shape of the box plot is Second quartile or 50% or median.
* The third black horizontal line of rectangle shape of the same box plot is third quartile or 75%
* The top black horizontal line of the rectangle shape of the box plot is the maximum value.
* The small diamond shape of the box plot is outlier data.

### Violin Plot

Violin plots also like boxplots summarize numeric data over a set of categories. They are essentially a box plot with a kernel density estimate (KDE) overlaid along with the range of the box and reflected to make it look nice. Unlike a box plot, in which all of the plot components correspond to actual data points, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x=df["average_count"])
```

```python
sns.violinplot(x="sector", y = "average_count", data = df)
```

## Relational Plot
There is the `relplot()` method which is used as the generic form of relational plot, within which you can call the `scatterplot()` (the default) or the `lineplot()`. For this session, we will focus on the `scatterplot()`. For more details on these plots, check out the documentation at https://seaborn.pydata.org/api.html (for the APIs) and https://seaborn.pydata.org/tutorial/relational.html (for the tutorial)

Show relationships between numerical variables.

```python
plt.figure(figsize = (10,5))

sns.scatterplot(x = 'loan_amount', y = 'lender_count', data = df)

plt.show()
```

We can also include hue parameter to add another dimension to our plot

Pass in sector to `hue` to get markers(data points) coloured differently depending on sector...

```python
plt.figure(figsize = (15,10))

sns.scatterplot(x = 'loan_amount', y = 'lender_count', data = df, hue = 'sector')

plt.show()
```

Pass in borrower genders to `style` to get markers style differentiated by gender, loan amount to `size` to get marker size differentiated by the loan amount, `sizes` to specify how big we want the markers to be (with a lower and upper limit for the size) and a value less than 1 to `alpha` to ensure transparency of markers overalying others.

```python
plt.figure(figsize = (15,10))

sns.scatterplot(x = 'loan_amount', y = 'lender_count', data = df, hue = 'sector',
               style = 'gender_lead',size = 'loan_amount',sizes = (100,500), alpha = 0.5)
#alpha is for transparency
plt.show()
```

We can use the parent function `relplot()`  together with `col` and `row` parameters to separate plots by categorical classes. (instead of overloading semantics on one plot)

We use the `col` and `row` parameters to generate subplots on columns and rows of the canvas, same in number as the categorical classes in the specified columns. 

`aspect` changes the aspect ratio of the plots for better legibility.

We want to have different plots showing the different categories of repayment interval.

```python
# We can have rows arranged columnwise by using col or 
sns.relplot(x = 'loan_amount', y = 'lender_count', data = df,
           col = 'repayment_interval', aspect = 1)
#col is split column wise

plt.show()
```

```python
# We can have rows arranged columnwise by using col or 
sns.relplot(x = 'loan_amount', y = 'lender_count', data = df,
           col = 'gender_lead', aspect = 1)
#col is split column wise

plt.show()
```

```python
# We can have rows arranged rowise by using row parameter
sns.relplot(x = 'loan_amount', y = 'lender_count', data = df,
           row = 'repayment_interval', aspect = 1.5)

plt.show()
```

We can use both row and col to include two extra variables.One will be plotted columnwise and the other row-wise

```python
sns.relplot(x = 'loan_amount', y = 'lender_count', data = df,
           row = 'gender_lead', col = 'repayment_interval', aspect = 1.5, hue = 'sector', size = 'loan_amount', sizes = (100,500))

plt.show()
```

## Distribution Plots
Of three types, `histplot()`, `kdeplot()` and `rugplot()`. We focus on the `distplot()`, which is the parent function and by default results in a histogram fitted with a kernel density estimate (outside the scope of this class. In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

To learn more about it, check out https://en.wikipedia.org/wiki/Kernel_density_estimation). For the other kinds, check out https://seaborn.pydata.org/api.html (for the APIs) and https://seaborn.pydata.org/tutorial/distributions.html (for the tutorial).

We can use the parent function distplot to plot either of these plots.

```python
sns.distplot(df['average_count'])

plt.show() 
```

Histogram

```python
sns.distplot(df['average_count'], kde = False)

plt.show()
```

```python
sns.distplot(df['average_count'], kde = False, bins = 10)

plt.show()
```

As you can see, they are all indistinguishably bunched up in the middle, but on either ends of the plot you can make out individual ticks.The center has higher density than the edges because most of the data occur there as opposed to the egdes

## Pair Plot
Visualizes the relationships between every numerical variable in the data frame.

Pairplots are used to generate a grid of plots of every numerical variable in the dataframe. Each column is compared to every other column (including itself). The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

This is good for a quick check to see what interesting relationships may exist between combinations of variables. Those that show promise can then be visualized using other tools to learn more from their relationship. For more on pairplots, check out the API at https://seaborn.pydata.org/generated/seaborn.pairplot.html#seaborn.pairplot.

```python
sns.pairplot(census)

plt.show()
```

## Matrix Plots
The last type of plots we shall look at are the matrix plots. These are the `sns.heatmap()` and the `sns.clustermap()`. We shall just focus on the `sns.heatmap()`. To find out more, check out the documentation at https://seaborn.pydata.org/generated/seaborn.heatmap.html#seaborn.heatmap (for the APIs).

We use the heatmaps to visualize correlation value of the different numerical variables.

To generate a heatmap, we need to generate a table of correlations from the dataframe.

```python
df_corr = census.corr()
df_corr
```

From above we see the correlation of the various numerical variables with each other. As expected the diagonals all show perfect correlation, because it is each variable correlated to itself. But in the other columns, we see high correlation between population and households, population and density and households and density.

We use the heatmap to get a visual of these correlations as it is easier to see at a glance based on colour difference the correlation between variables.

```python
sns.heatmap(df_corr)

plt.show()
```

Importance of this will come out in ML's feature engineering.

In the above plot, the darker the box, the lower the correlation and vice versa.

We can specify what colour scheme to use. Below we use `coolwarm`, with warm being high correlation and cool being low correlation.

Check out this page from the Matplotlib documentation to see what colour mappings you can use: https://matplotlib.org/tutorials/colors/colormaps.html

```python
sns.heatmap(df_corr, cmap = 'coolwarm')

plt.show()
```

We can also annotate the boxes with the actual correlation values setting the `annot` parameter to `True`. We can also add a border between the boxes by setting the `linewidth` parameter to an appropriate value.

```python
sns.heatmap(df_corr, cmap = 'YlGnBu', annot = True, linewidth = 0.5)

plt.show()
```
