# How to deal with missing values

there are various ways of handling this data. one can either reeplace null values with real values like mean, mode median

depending with the dayta that you are working on, or you can just drop the rows with missing values. each has its 

pros and cons

**drop missing values**

if you want to simply exclude the missing values, then use the dropna() function along with the axis argument.

by default `axis = index` , ie along rows which means that if any value within a row is na then the whole row is excluded.

we can also drop columns by indicating axis is columns.

```python
data.dropna()
```

```python
data.dropna()
```

```python
data.dropna(axis='rows',inplace=True)
```

```python
data
```

```python
data.dropna(axis='columns')
```

```python
data.dropna(axis=1)# 1 is the same as the columns(drops the columns)
```

```python
data.dropna(0)# 0 = rows
```

   **pros**

- complete removal of data with missing values result in robust and highly accurate model

- deleting a particular row or column with no specific info is better since it does not have a high sabotage.

  **cons**


- loss of information and data

- works poorly in the percentage of missing values in high(say 30%) compared to the whole dataset.

# boolean subsetting

**You'll learn how to:**

- Filter and segment data using boolean indexing


In boolean indexing, we will select subsets of data based on the actual values of the data in the DataFrame and not on their row/column labels or integer locations. In boolean indexing, we use a boolean vector to filter the data

Boolean indexing is a type of indexing that uses actual values of the data in the DataFrame. In boolean indexing, we can filter a data in four ways:

- Accessing a DataFrame with a boolean index
- Applying a boolean mask to a dataframe
- Masking data based on column value
- Masking data based on an index value

# Merging

when merging two datasets together, it is usually important to merge the datasets on the column that exists in

both datasets and it has unique values. For instance, we merged our data and area datasets on `Subcounty` 

since it is unique. This helps in avoiding to re-do the cleaning again on the merged dataframe if the merging was

done incorrectly. 

code syntax: pd.merge(dataframe_A, determine_B, on='column_name')

```python
data=pd.read_excel('census population households.xlsx', thousands=',')
data
```

```python
#check for missing values
#replace this values with the median
```

```python
#check for missing values
data.isnull().sum()
```

```python
data['Population'].fillna(data['Population'].median(), inplace=True)
```

```python
data['Households'].fillna(data['Households'].median(), inplace=True)
```

```python
data['Average Household Size'].fillna(data['Average Household Size'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(subset='Sub County', inplace=True)
```
