## Solution

```python
#import pandas
import pandas as pd

#read data
fp = ('6153237444115dat.csv')

data = pd.read_csv(fp,
                   na_values = ['*', '**', '***', '****', '*****', '******'],
                   usecols = ['USAF', 'YR--MODAHRMN', 'TEMP', 'MAX', 'MIN'],
                   parse_dates = ['YR--MODAHRMN'], #converts date to date time
)

#rename columns
data.rename(columns={'YR--MODAHRMN':'DATETIME', 'USAF':'STATION'}, inplace=True)

# Create column from datetime for yr-mn-dt
data['DATE'] = data['DATETIME'].dt.date

#drop na values
data= data.dropna(subset=['TEMP'])

# creates new column Celsius and adds fahrenheits converted to celsius in it
data['TEMP_C'] = (data["TEMP"] - 32) / 1.8
# round values in column Celsius to integer
data['TEMP_C'] = data['TEMP_C'].round(2).astype(int)

#create groups by day
day_grouped = data.groupby('DATE')

#Get mean, max and min values
mean_val = day_grouped['TEMP_C'].mean()
max_val = day_grouped['TEMP_C'].max()
min_val = day_grouped['TEMP_C'].min()

#add them to a new dataframe and rename the columns
daily_data = pd.concat([mean_val, max_val, min_val], axis=1)
daily_data.columns = ['Daily Mean', 'Daily Max', 'Daily Min']
daily_data
```

## For Kumpula and Rovaniem Data

```python
#import pandas
import pandas as pd

#Read in the data
kumpula = pd.read_csv('Kumpula_temps_May_Aug_2017.csv',
                      parse_dates = ['YR--MODAHRMN']
                     )

rovaniemi = pd.read_csv('Rovaniemi_temps_May_August_2017.csv',
                        parse_dates = ['YR--MODAHRMN']
                       )
#delete columns
del(kumpula['Unnamed: 0'])
del(rovaniemi['Unnamed: 0'])

#add date column
kumpula['DATE'] = kumpula['YR--MODAHRMN'].dt.date
rovaniemi['DATE'] = rovaniemi['YR--MODAHRMN'].dt.date

#group by date
kumpula_date_group = kumpula.groupby('DATE')
rovaniemi_date_group = rovaniemi.groupby('DATE')

#calculate min, max and mean
kumpula_mean = kumpula_date_group['Celsius'].mean()
kumpula_max = kumpula_date_group['Celsius'].max()
kumpula_min = kumpula_date_group['Celsius'].min()

rovaniemi_mean = rovaniemi_date_group['Celsius'].mean()
rovaniemi_max = rovaniemi_date_group['Celsius'].max()
rovaniemi_min = rovaniemi_date_group['Celsius'].min()

#add to new data frame
kumpula_data = pd.concat([kumpula_mean, kumpula_max, kumpula_min], axis=1)
kumpula_data.columns = ['Kumpula Mean', 'Kumpula Max', 'Kumpula Min']

rovaniemi_data = pd.concat([rovaniemi_mean, rovaniemi_max, rovaniemi_min], axis=1)
rovaniemi_data.columns = ['Rovaniemi Mean', 'Rovaniemi Max', 'Rovaniemi Min']
```
